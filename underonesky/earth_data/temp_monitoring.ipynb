## Extract temp data from log file

```python
import pandas as pd
with open('underonesky_supervisor.log', 'r') as fr:
    # reading line by line
    lines = fr.readlines()

    # opening in writing mode
    with open('underonesky_supervisor_temp.log', 'w') as fw:
        fw.write("time|cpu|temp1|temp2\n")
        for line in lines:
            if "TEMP" in line:
                line = line.replace(" TEMP(CPU 1 2)", "")
                fw.write(line)
df = pd.read_csv("underonesky_supervisor_temp.log", delimiter="|", parse_dates=True)
df = df.set_index("time")
df
```

## Plot results

```python
df.plot(rot=45)
```
